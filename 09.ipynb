--- Day 9: Rope Bridge ---

This rope bridge creaks as you walk along it. You aren't sure how old it is, or whether it can even support your weight.

It seems to support the Elves just fine, though. The bridge spans a gorge which was carved out by the massive river far below you.

You step carefully; as you do, the ropes stretch and twist. You decide to distract yourself by modeling rope physics; maybe you can even figure out where not to step.

Consider a rope with a knot at each end; these knots mark the head and the tail of the rope. If the head moves far enough away from the tail, the tail is pulled toward the head.

Due to nebulous reasoning involving Planck lengths, you should be able to model the positions of the knots on a two-dimensional grid. Then, by following a hypothetical series of motions (your puzzle input) for the head, you can determine how the tail will move.

Due to the aforementioned Planck lengths, the rope must be quite short; in fact, the head (H) and tail (T) must always be touching (diagonally adjacent and even overlapping both count as touching):



```
....
.TH.
....

....
.H..
..T.
....

...
.H. (H covers T)
...
```



If the head is ever two steps directly up, down, left, or right from the tail, the tail must also move one step in that direction so it remains close enough:



```
.....    .....    .....
.TH.. -> .T.H. -> ..TH.
.....    .....    .....

...    ...    ...
.T.    .T.    ...
.H. -> ... -> .T.
...    .H.    .H.
...    ...    ...
```



Otherwise, if the head and tail aren't touching and aren't in the same row or column, the tail always moves one step diagonally to keep up:



```
.....    .....    .....
.....    ..H..    ..H..
..H.. -> ..... -> ..T..
.T...    .T...    .....
.....    .....    .....

.....    .....    .....
.....    .....    .....
..H.. -> ...H. -> ..TH.
.T...    .T...    .....
.....    .....    .....
```

````
record Point(int X, int Y)
{
    public static Point Zero = new Point(0,0);
}

record Rope(Point H, Point T);

Point Move(Point p, string direction)
{
    var x = direction switch
    {
        string d when d.Contains("R") => 1,
        string d when d.Contains("L") => -1,
        _ => 0       
    };

    var y = direction switch
    {
        string d when d.Contains("U") => 1,
        string d when d.Contains("D") => -1,
        _ => 0       
    };

    var start = p;
    return new Point(p.X + x, p.Y + y);
}

IEnumerable<Point> Move(Point p, string direction, int count)
{
    var loc = p;
    for (int i=0; i<count; i++)
    {
        loc = Move(loc,direction);
        yield return loc;
    } 
}


var moves = new List<Point>() { new Point(0,0) };

var positions = commands.Aggregate(moves, (agg, val) => 
{
    agg.AddRange(Move(agg.Last(), val.direction, val.count));
    return agg;
});

display(positions);
````

````
record LongRope(Point[] Pieces)
{
    public LongRope(int length) : this(new Point[length]) {}
    public Point H => Pieces.First();
    public Point T => Pieces.Last();
    public static LongRope Init(Point p, int length)
    {
        var result = new LongRope(length);
        for (int i=0; i< length; i++) result.Pieces[i] = p;
        return result; 
    }
    public static LongRope DeepClone(LongRope r)
    {
        var result = new LongRope(r.Pieces.Length);
        r.Pieces.CopyTo(result.Pieces,0);
        return result; 
    }
}

LongRope Move(LongRope oldRope, string direction)
{
    var r = LongRope.DeepClone(oldRope);
    var head = r.H;
    r.Pieces[0] = Move(head, direction);
    for (int i=1; i < r.Pieces.Length; i++)
    {
        var lead = r.Pieces[i-1];
        var follow = r.Pieces[i];
        (bool touching, int XSign, int YSign) = MoveInfo(follow, lead);
        if (touching) break;
        r.Pieces[i] = new Point(follow.X + XSign, follow.Y + YSign);
    }

    return r;
}

IEnumerable<LongRope> Move(LongRope r, string direction, int count)
{
    var loc = r;
    for (int i=0; i<count; i++)
    {
        loc = Move(loc,direction);
        yield return loc;
    } 
}

String Render(LongRope r)
{
    StringBuilder sb = new();
    var minX = Math.Min(-10,r.Pieces.Select(p => Math.Min(p.X, p.X)).Min());
    var maxX = Math.Max(10,r.Pieces.Select(p => Math.Max(p.X, p.X)).Max());
    var minY = Math.Min(-10,r.Pieces.Select(p => Math.Min(p.Y, p.Y)).Min());
    var maxY = Math.Max(10,r.Pieces.Select(p => Math.Max(p.Y, p.Y)).Max());
    for (int y = maxY; y>=minY; y--)
    {
        for (int x = minX; x<=maxX; x++)
        {
            var p = new Point(x,y);
            char c = p == Point.Zero ? 's' : '.';

            foreach (var piece in r.Pieces)
            {
                if (c != '.') continue;

                if (piece == p) c = '#';
                else if (piece == p) c = '#';
            }
            if (r.T == p) c = 'T';
            else if (r.H == p) c = 'H';

            sb.Append(c);
        }
        sb.AppendLine();
    }
    return sb.ToString();
}

var longRopeMoves = new List<LongRope>() { LongRope.Init(new Point(0,0), 10) };

// var header = display("Ready");
// var image = display("Ready");

var longRopePositions = commands.Aggregate(longRopeMoves, (agg, val) => 
{
    var moves = Move(agg.Last(), val.direction, val.count).ToList();

    // foreach (var move in moves)
    // {
    //     header.Update(val);
    //     image.Update(Render(move));
    //     System.Threading.Thread.Sleep(100);
    // }

    agg.AddRange(moves);
    return agg;
});


var distinctLongTailPositions = longRopePositions.Select(r => r.T).Distinct();

Console.WriteLine($"Day 9 Part 2 : {distinctLongTailPositions.Count()}");
````
